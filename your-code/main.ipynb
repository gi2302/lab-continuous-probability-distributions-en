## Uniform Distribution

To generate uniform random numbers between any two given values using scipy, we can either use the following code or the code that we have
discussed in class:

```python
# your code here

from scipy.stats import uniform

# Function to generate uniformly distributed random numbers using SciPy
def generate_uniform(bottom, ceiling, count):
    return uniform.rvs(loc=bottom, scale=ceiling - bottom, size=count)
```

```python
# your code here

# Perform the Shapiro-Wilk test for normality (for each variable)
from scipy.stats import shapiro

# Sample size limit (you can adjust this based on your needs, e.g., 5000 or smaller)
sample_size = 5000

# Shapiro-Wilk test for normality on a random sample to avoid the warning
sample_fuel = vehicles_df['Fuel Barrels/Year'].dropna().sample(n=sample_size, random_state=42)
sample_co2 = vehicles_df['CO2 Emission Grams/Mile'].dropna().sample(n=sample_size, random_state=42)
sample_mpg = vehicles_df['Combined MPG'].dropna().sample(n=sample_size, random_state=42)

# Shapiro-Wilk Test
stat_fuel, p_fuel = shapiro(sample_fuel)
stat_co2, p_co2 = shapiro(sample_co2)
stat_mpg, p_mpg = shapiro(sample_mpg)

print(f"Shapiro-Wilk Test for Fuel Barrels/Year: p-value = {p_fuel}")
print(f"Shapiro-Wilk Test for CO2 Emission Grams/Mile: p-value = {p_co2}")
print(f"Shapiro-Wilk Test for Combined MPG: p-value = {p_mpg}")
```

### Shapiro-Wilk Test for Normality:

The Shapiro-Wilk test is a statistical test used to determine whether a variable follows a normal distribution. The null hypothesis for this test is that the data is normally distributed. A **p-value** less than 0.05 indicates that we can reject the null hypothesis, suggesting that the data is not normally distributed.

The results for the Shapiro-Wilk test for "Fuel Barrels/Year," "CO2 Emission Grams/Mile," and "Combined MPG" are as follows:

- **Fuel Barrels/Year**: p-value = 2.55e-33
- **CO2 Emission Grams/Mile**: p-value = 3.20e-33
- **Combined MPG**: p-value = 2.20e-35

### Interpretation:
- Since all three p-values are extremely small (much less than 0.05), we can **reject the null hypothesis** for all three variables. This indicates that none of the variables follow a normal distribution.
- These results align with the previous visual analyses (histograms, boxplots, and Q-Q plots), which showed signs of skewness and non-normality.

### Conclusion:
The Shapiro-Wilk test confirms that the data for "Fuel Barrels/Year," "CO2 Emission Grams/Mile," and "Combined MPG" is **not normally distributed**. The p-values suggest significant departures from normality for all three variables.

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
import numpy as np
import matplotlib.pyplot as plt

# Function to generate exponentially distributed random numbers
def generate_exponential(mean, size):
    return np.random.exponential(mean, size)

# Generate two sequences
data_10 = generate_exponential(10, 10)
data_100 = generate_exponential(10, 100)

# Plot the distributions as histograms
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(data_10, bins=100, edgecolor='black')
plt.title('Exponential Distribution (size=10)')
plt.xlabel('Value')
plt.ylabel('Count')

plt.subplot(1, 2, 2)
plt.hist(data_100, bins=100, edgecolor='black')
plt.title('Exponential Distribution (size=100)')
plt.xlabel('Value')
plt.ylabel('Count')

plt.tight_layout()
plt.show()

# Explanation
print("Note: The outputs differ each time because the random numbers are generated randomly, so the distribution will vary with each run. This is expected behavior for random processes.")
```

How are the two distributions different?

### Answer:

The two distributions generated by the exponential function differ in the following ways:

1. **Size of the Data**:
   - The first distribution, with **size=10**, is more discrete, as indicated by the high frequency counts in specific bins. This is expected because with fewer data points, each value has a greater influence, making the distribution appear more "spiky."
   - The second distribution, with **size=100**, has a smoother appearance and a more continuous pattern. With more data points, the distribution becomes less erratic and follows the exponential distribution's expected shape more clearly.

2. **Distribution Shape**:
   - For the **size=10**, the values are concentrated in the lower range (near zero), with several bins having a count of 1. This is because fewer samples mean less variation and a higher concentration of lower values.
   - The **size=100** distribution is more spread out, with the data values covering a broader range. The pattern is more typical for an exponential distribution, with a rapid drop in frequency as the value increases.

3. **Statistical Stability**:
   - As we increase the sample size, the distribution starts to resemble the expected **exponential distribution** more closely. In contrast, with smaller sample sizes, the data can have more variance and may not show the typical distribution shape.

### Summary:
The distribution with **size=100** shows a smoother and more predictable exponential decay, while the **size=10** distribution is more discrete and erratic. The larger sample size helps reveal the true nature of the exponential distribution.

## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e. λ = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
# your code here
# Hint: This is same as saying P(x<15)

import numpy as np
from scipy.stats import expon

# Parameters
mean = 10  # Mean time in minutes
lambda_param = 1 / mean  # λ (rate)

# Calculate the probability that a customer spends less than 15 minutes
x = 15
prob_less_than_15 = expon.cdf(x, scale=1/lambda_param)

print(f"Probability that a customer spends less than 15 minutes in the bank: {prob_less_than_15:.4f}")
```

What is the probability that the customer will spend more than 15 minutes

```python
# your code here

# Calculate the probability that a customer spends more than 15 minutes
prob_more_than_15 = 1 - prob_less_than_15

print(f"Probability that a customer spends more than 15 minutes in the bank: {prob_more_than_15:.4f}")
```
